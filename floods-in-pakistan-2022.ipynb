# Visualization

# How many people died due to flood in each Region

```python
import plotly.express as px

fig = px.bar(pak_flood, x="Region", y=["M_D", "F_D", "C_D"], title="Deaths per Region", 
             text_auto=True, width=900, height=700)
fig.show()
```

# Total Injured people per region

```python
import plotly.express as px

fig = px.bar(pak_flood, x="Region", y=["M_I", "F_I", "C_I"], title="Injured per Region", 
             text_auto=True, width=900, height=700)
fig.show()
```

# Impacts of Flood on livestock, population, houses

```python
import plotly.express as px

fig = px.bar(pak_flood, x="Region", y=["Total_houses_damaged", "Affected population", "livestock_damaged"], title="Impacts of Flood", 
             text_auto=True, width=900, height=700)
fig.show()
```

```python
import plotly.express as px

fig = px.bar(pak_flood, x="Region", y=["Roads_damaged(km)", "Bridges damaged"], title="Impacts of Flood on Roads and Bridges", 
             text_auto=True, width=900, height=700)
fig.show()
```

# Impacts of Pakistan Flood 2022 on Various Areas

```python
import plotly.express as px

fig = px.bar(pak_flood, x="Region", y=["Roads_damaged(km)", "Bridges damaged", "Total_houses_damaged", "Affected population", "livestock_damaged", "total_injured", "Total deaths"], title="Impacts of Flood on Various Areas", 
             text_auto=True, width=900, height=700)
fig.show()
```
